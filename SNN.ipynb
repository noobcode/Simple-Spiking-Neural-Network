```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Izhikevic model

```python
class Izhikevic_model:
    def __init__(self, a=0.02, b=0.2, c=-65, d=8, n_neurons=1, neurons={}, neuron_type=''):
        # initialize paramiters
        self.set_parameters(neuron_type, a, b, c, d)
        # initialize synaptic weights
        self.S = 0.5 * np.random.randn(n_neurons,n_neurons)
        # initialize voltages and recoveries
        self.v = c * np.ones(shape=(n_neurons,))
        self.u = b * self.v
        self.n_neurons = n_neurons
        
    def set_parameters(self, neuron_type, a, b, c, d):
        if neuron_type == '':
            self.a, self.b, self.c, self.d = a, b, c, d
        if neuron_type == 'rs': # regular spiking
            self.a, self.b, self.c, self.d = 0.02, 0.2, -65, 8
        if neuron_type == 'ib': # intrinsically bursting
            self.a, self.b, self.c, self.d = 0.02, 0.2, -55, 4
        if neuron_type == 'ch': # chattering
            self.a, self.b, self.c, self.d = 0.02, 0.2, -50, 2
        if neuron_type == 'fs': # fast spiking
            self.a, self.b, self.c, self.d = 0.1, 0.2, -65, 2
        if neuron_type == 'tc': # thalamo-cortical
            self.a, self.b, self.c, self.d = 0.02, 0.25, -65, 0.05
        if neuron_type == 'rz': # resonator
            self.a, self.b, self.c, self.d = 0.1, 0.26, -65, 2
        if neuron_type == 'lts': # low-threshold spiking
            self.a, self.b, self.c, self.d = 0.02, 0.25, -65, 2
            
        
    def simulate(self, simulation_time, delta_t, input_type=""):
        # variables to save history of firings, voltages and recoveries
        firings = np.empty((0,2))
        voltages = np.empty(shape=(self.n_neurons, simulation_time))
        recoveries = np.empty(shape=(self.n_neurons, simulation_time))

        # simulate
        for t in range(simulation_time):
            I = 10 * np.ones(self.n_neurons) #np.random.randn(self.n_neurons)
            # if neuron has fired
            has_fired = np.nonzero(self.v >= 30)
            self.v[has_fired] = 30 # clip voltage to apex
            tmp_v = np.copy(self.v)
            self.v[has_fired] = self.c # after-spike resetting
            self.u[has_fired] = self.u[has_fired] + self.d
    
            # otherwise
            I += np.sum(self.S[has_fired[0]], axis=0)
            step = int(1/delta_t)
            for s in range(step):
                # compute 1 ms of v in 1/dt steps to avoid numerical problems
                self.v = self.v + (0.04 * np.square(self.v) + 5 * self.v + 140 - self.u + I) * delta_t
                self.u = self.u +  self.a * (self.b * self.v - self.u) * delta_t
    
            # save results
            if len(has_fired[0]) > 0:
                fired_t = [t] * len(has_fired[0])
                firings = np.vstack((firings, np.vstack((fired_t, has_fired[0])).T))
            voltages[:, t] = tmp_v
            recoveries[:, t] = self.u
        
        return firings, voltages, recoveries
```

#### utility plot functions

```python
def plot_neuron_voltages_and_recoveries(idx, voltages, recoveries):
    plt.plot(voltages[idx, :], label="voltage", color='black')
    plt.plot(recoveries[idx, :], label="recovery", color='red')
    plt.axhline(y=30, color='green', linestyle='--', label="AP apex")
    plt.xlabel("time (ms)")
    plt.ylabel("membrane potential (mV)")
    plt.legend(loc='best')
    plt.show()
```

```python
def spike_raster_plot(firings):
    plt.scatter(firings[:, 0], firings[:, 1], marker="_", label="spike", alpha=1, color='black')
    plt.ylabel("neuron number")
    plt.xlabel("time (ms)")
    plt.legend(loc='best')
    plt.show()
```

```python
def spikes_overtime(firings):
    # aggregate firings by time
    unique_timings = np.unique(firings[:,0])
    spike_numbers = [ len(firings[firings[:, 0] == t]) for t in unique_timings]
    for t, n in zip(unique_timings, spike_numbers):
        plt.plot([t, t], [0, n], linewidth=6, alpha=0.6, color='black')
    
    plt.xlabel('time (ms)')
    plt.ylabel('# spikes')
    plt.show()
```

## Network simulation
- membrane and recovery plot of a neuron simulation
- spike raster plot of a network of spiking neurons simulation
- histogram of spikes

```python
net = Izhikevic_model(n_neurons=1000, neuron_type='rs')
firings, voltages, recoveries = net.simulate(simulation_time=200, delta_t=0.5, input_type="")

neuron_idx = 5
plot_neuron_voltages_and_recoveries(neuron_idx, voltages, recoveries)
spike_raster_plot(firings)
spikes_overtime(firings)
```
